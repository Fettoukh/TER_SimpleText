```python
import requests
import json
```

```python
import elasticsearch
es = elasticsearch.Elasticsearch()

res = es.search(
    index="test2", 
    body={"query": {"match_all": {}}, "size": 10000, "fields": ["_id"]})

ids = [d['_id'] for d in res['hits']['hits']]
```

```python
for id in ids:
    myobj = {
    "docs" : [
        {
        "_id" : id,
        "fields": ["article.body.text"],
        "term_statistics" : "true"
        }
    ]}
```

```python
url = 'http://localhost:9200/test2/_mtermvectors'

x = requests.post(url, data = json.dumps(myobj), headers=headers)

doc_freqs = []
term_freqs = []
mots = []
stats=[]
```

```python
json_data = x.json()
print(json_data)
```

```python
#print the response text (the content of the requested file):
   # abstract = doc["term_vectors"]["article.front.article-meta.author-notes.fn.p"]


    
json_data = x.json()
#print(json_data)
#print(json_data["docs"][0]["term_vectors"]["article.front.article-meta.author-notes.fn.p"])

for doc in json_data["docs"]:
    abstract = doc["term_vectors"]["article.body.text"]
    # print(doc["term_vectors"].keys())
    terms = abstract["terms"]
    # print(terms.values())
    for term in terms.values():
        doc_freqs.append(term["doc_freq"])
        term_freqs.append(term["term_freq"])
        # print(" doc_freq =",term["doc_freq"])
        # print(" term_freq =", term["term_freq"])
    # print(doc_freqs)
    for term in terms.keys():
        mots.append(term)
    #     print("term =",term)

    for x in range(len(mots)):
        # print(mots[x])
        stats.append({"term" : mots[x] , "doc_freq" : doc_freqs[x] , "term_freq" : term_freqs[x]})
    print(stats)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
pd.set_option("max_rows", 600)
from pathlib import Path  
import glob
```

```python
directory_path = "C://Users/wafae/Desktop/New-folder"
```

```python
text_files = glob.glob(f"{directory_path}/*.txt")
```

```python
text_files
```

```python
text_titles = [Path(text).stem for text in text_files]
text_titles
```

```python
tfidf_vectorizer = TfidfVectorizer(input='filename', stop_words='english')
```

```python
tfidf_vector = tfidf_vectorizer.fit_transform(text_files)
```

```python
tfidf_df = pd.DataFrame(tfidf_vector.toarray(), index=text_titles, columns=tfidf_vectorizer.get_feature_names())
tfidf_df
```

```python
tfidf_df.loc['00_Document Frequency'] = (tfidf_df > 0).sum()
```
