```python
# !pip install elasticsearch
# !pip install wikipedia

import requests
import json
import elasticsearch
import math
import operator
import pandas as pd
```

```python
# Get Documents IDs

ids_url = 'http://localhost:9200/test3/_search'

x = requests.get(ids_url)
json_data = x.json()
ids = [d['_id'] for d in json_data['hits']['hits']]
ids
```

```python
# get DocFreq and termFreq of each term in our collection of documents

headers = {"Content-Type": "application/json"}
mtermvectors_data = {}
for id in ids:
    myobj = {
    "docs" : [
        {
        "_id" : id,
        "fields": ["article.body.text"],
        "term_statistics" : "true"
        }
    ]}
#     print(id)
    mtermvectors_url = 'http://localhost:9200/test3/_mtermvectors'
    x = requests.post(mtermvectors_url, data = json.dumps(myobj), headers=headers)
    mtermvectors_data.update(x.json()["docs"][0]["term_vectors"]["article.body.text"]["terms"]) 
print(mtermvectors_data)
```

```python
# organizing data in a single list

doc_freqs = []
term_freqs = []
terms = []
all_words=[]
# freqs = 0


for term in mtermvectors_data.values():
    doc_freqs.append(term["doc_freq"])
    term_freqs.append(term["ttf"])

for term in mtermvectors_data.keys():
    terms.append(term)

for x in range(len(terms)):
    all_words.append({"term" : terms[x] , "doc_freq" : doc_freqs[x] , "term_freq" : term_freqs[x]})
    
print(len(all_words))
```

```python
# Removing common words from our list

english_most_common_10k = 'https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears.txt'
english_most_common_10k_long = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-long.txt"
english_most_common_10k_medium = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-medium.txt"
english_most_common_10k_shorts = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-short.txt"

dictionaries_urls = [english_most_common_10k,english_most_common_10k_long,english_most_common_10k_medium ,english_most_common_10k_shorts]

def getRequest(url):
    response = requests.get(url)
    data = response.text 
    return data


common_words = ""
for url in dictionaries_urls:
    x = getRequest(url)
    common_words+=x
      
set_of_common_words = common_words.split('\n')

# print(set_of_common_words)
set_without_common_words = []
for x in range(len(terms)):
    if(all_words[x]["term"] not in set_of_common_words):
        set_without_common_words.append({"term" : all_words[x]["term"] , "doc_freq" : all_words[x]["doc_freq"] , "term_freq" : all_words[x]["term_freq"]})

# for x in range(len(set_without_common_words)):
#     print(set_without_common_words[x]["term"])       
```

```python
#Get the number od documents in our collection

urlCountDoc = 'http://localhost:9200/test3/_count'
myobj = {}

countDoc = requests.post(urlCountDoc, data = json.dumps(myobj), headers=headers)
json_countDoc = countDoc.json()
nbr_docs = json_countDoc["count"]
print(nbr_docs)
```

# IDF Filtering


```python
# Calculate IDF and filtering the terms

hardWords_IDF = []
for x in range(len(set_without_common_words)):

    term_IDF = math.log10( 1 + (nbr_docs/set_without_common_words[x]["doc_freq"]))
#     print(set_without_common_words[x]["term"] , term_IDF)
    if(len(set_without_common_words[x]["term"])>=4 and term_IDF>0.6 ):
        hardWords_IDF.append({"term" : set_without_common_words[x]["term"] , "doc_freq" : set_without_common_words[x]["doc_freq"] , "term_freq" : set_without_common_words[x]["term_freq"], "idf" : term_IDF})
len(hardWords_IDF)

```

# BM25 Filtering

```python
# Calculate IDF With BM25 Algo


hardWordsBM25 = []
for x in range(len(set_without_common_words)):

    up = nbr_docs - set_without_common_words[x]["doc_freq"] + 0.5
    down = set_without_common_words[x]["doc_freq"] + 0.5
    BM25_IDF = math.log10(1+(up/down))
#     print(set_without_common_words[x]["term"] , BM25_IDF)
    if(len(set_without_common_words[x]["term"])>5 and BM25_IDF>0.5 ):
        hardWordsBM25.append({"term" : set_without_common_words[x]["term"] , "doc_freq" : set_without_common_words[x]["doc_freq"] , "term_freq" : set_without_common_words[x]["term_freq"], "idf" : BM25_IDF})
print(len(hardWordsBM25))
```

```python
# Calculate IDF and filtering the terms

hardWords_IDF_Freq = []
for x in range(len(set_without_common_words)):

    term_IDF = math.log10( nbr_docs/set_without_common_words[x]["doc_freq"] + set_without_common_words[x]["term_freq"])
#     print(set_without_common_words[x]["term"], set_without_common_words[x]["term_freq"])
#     print(set_without_common_words[x]["term"] , term_IDF)
    if(len(set_without_common_words[x]["term"])>=4 and term_IDF>0.7):
        hardWords_IDF_Freq.append({"term" : set_without_common_words[x]["term"] , "doc_freq" : set_without_common_words[x]["doc_freq"] , "term_freq" : set_without_common_words[x]["term_freq"], "idf" : term_IDF})
len(hardWords_IDF_Freq)
```
