```python
# !pip install elasticsearch
# !pip install wikipedia

import requests
import json
import elasticsearch
import math
import operator
import pandas as pd
```

```python
# Get Documents IDs

url = 'http://localhost:9200/test3/_search'

x = requests.get(url)
json_data = x.json()
ids = [d['_id'] for d in json_data['hits']['hits']]
ids
```

```python
# get DocFreq and termFreq of each term in our collection of documents

headers = {"Content-Type": "application/json"}
json_data = {}
for id in ids:
    myobj = {
    "docs" : [
        {
        "_id" : id,
        "fields": ["article.body.text"],
        "term_statistics" : "true"
        }
    ]}
#     print(id)
    url = 'http://localhost:9200/test3/_mtermvectors'
    x = requests.post(url, data = json.dumps(myobj), headers=headers)
    json_data.update(x.json()["docs"][0]["term_vectors"]["article.body.text"]["terms"]) 
print(len(json_data.values()))
```

```python
# organizing data in a single list

doc_freqs = []
term_freqs = []
mots = []
stats=[]
freqs = 0


for term in json_data.values():
    doc_freqs.append(term["doc_freq"])
    term_freqs.append(term["term_freq"])

for term in json_data.keys():
    mots.append(term)

for x in range(len(mots)):
    stats.append({"term" : mots[x] , "doc_freq" : doc_freqs[x] , "term_freq" : term_freqs[x]})
    
print(len(stats))
```

```python
# Removing common words from our list

english_most_common_10k = 'https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears.txt'
english_most_common_10k_long = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-long.txt"
english_most_common_10k_medium = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-medium.txt"
english_most_common_10k_shorts = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears-short.txt"

urls = [english_most_common_10k,english_most_common_10k_long,english_most_common_10k_medium ,english_most_common_10k_shorts]

def getRequest(url):
    response = requests.get(url)
    data = response.text 
    return data


common_words = ""
for url in urls:
    x = getRequest(url)
    common_words+=x
      
set_of_common_words = common_words.split('\n')

# print(set_of_common_words)
set_without_common_words = []
for x in range(len(mots)):
    if(stats[x]["term"] not in set_of_common_words):
        set_without_common_words.append({"term" : stats[x]["term"] , "doc_freq" : stats[x]["doc_freq"] , "term_freq" : stats[x]["term_freq"]})

# for x in range(len(set_without_common_words)):
#     print(set_without_common_words[x]["term"])       
```

```python
#Get the number od documents in our collection

urlCountDoc = 'http://localhost:9200/test3/_count'
myobj = {}

countDoc = requests.post(urlCountDoc, data = json.dumps(myobj), headers=headers)
json_countDoc = countDoc.json()
N = json_countDoc["count"]
print(N)
```

```python
# Calculate IDF and filtering the terms

hardWords1 = []
for x in range(len(set_without_common_words)):

    term_IDF = math.log10( N/set_without_common_words[x]["doc_freq"])
#     print(set_without_common_words[x]["term"] , term_IDF)
    if(len(set_without_common_words[x]["term"])>=4 and term_IDF>0.6 ):
        hardWords1.append({"term" : set_without_common_words[x]["term"] , "doc_freq" : set_without_common_words[x]["doc_freq"] , "term_freq" : set_without_common_words[x]["term_freq"], "idf" : term_IDF})
# hardWords1

```

```python
# Removing terms that doesn t appear in wikipedia or wikitionary

hardWords2 = []

def getWiktionary(term):
    urlCountDoc = 'https://en.wiktionary.org/wiki/'+term
    myobj = {}

    response = requests.get(urlCountDoc, data = json.dumps(myobj), headers=headers)
    return response

import wikipedia as wiki
for x in range(len(hardWords1)):
    if (wiki.search(hardWords1[x]["term"]) != [] or getWiktionary(hardWords1[x]["term"]).status_code == 200) :
         hardWords2.append({"term" : hardWords1[x]["term"] , "doc_freq" : hardWords1[x]["doc_freq"] , "term_freq" : hardWords1[x]["term_freq"], "idf" : hardWords1[x]["idf"]})
```

```python
# Calculate IDF With BM25 Algo


hardWordsBM25 = []
for x in range(len(set_without_common_words)):

    up = N - set_without_common_words[x]["doc_freq"] + 0.5
    down = set_without_common_words[x]["doc_freq"] + 0.5
    BM25_IDF = math.log10(1+(up/down))
#     print(set_without_common_words[x]["term"] , BM25_IDF)
    if(len(set_without_common_words[x]["term"])>5 and BM25_IDF>0.7 ):
        hardWordsBM25.append({"term" : set_without_common_words[x]["term"] , "doc_freq" : set_without_common_words[x]["doc_freq"] , "term_freq" : set_without_common_words[x]["term_freq"], "idf" : BM25_IDF})
# print(len(ardWordsBM25)
```
